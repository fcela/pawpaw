# Classification Test Suite Diagnostics

This notebook diagnoses why test suites report 0% accuracy and demonstrates what the classifier actually does.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd()))
import pawpaw

classifier = pawpaw.load('./programs/triage')
print('Loaded triage classifier')
```

## What does this classifier do?

```python
for inp in ['Urgent: Server down!', 'FYI: Meeting at 3pm', 'Lunch?']:
    print(f'{inp} -> {classifier(inp)}')
```

## Why sentiment tests fail (0% accuracy)

```python
tests = [('I love this!', 'positive'), ('This is terrible', 'negative')]
for inp, expected in tests:
    result = classifier(inp)
    print(f'Input: {inp} | Expected: {expected} | Got: {result} | Match: {result==expected}')
```

## What it IS good at: Triage (trivial vs substantive)

```python
triage_tests = [
    ('Urgent: Server down!', 'substantive'),
    ('Lunch tomorrow?', 'trivial'),
    ('IMPORTANT: Security breach!', 'substantive')
]
correct = sum(1 for inp, exp in triage_tests if classifier(inp).lower() == exp.lower())
print(f'Triage accuracy: {correct}/{len(triage_tests)} ({correct/len(triage_tests)*100:.0f}%)')
```
